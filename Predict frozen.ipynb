```python
import os
os.environ['TF_FORCE_GPU_ALLOW_GROWTH'] = 'true'
```

```python
import tensorflow as tf
from tensorflow.python.platform import gfile
```

```python
f = gfile.FastGFile('model/fast_scnn.pb', 'rb')
```

```python
graph_def = tf.GraphDef()
```

```python
graph_def.ParseFromString(f.read())
```

```python
f.close()
```

```python
sess = tf.Session()
```

```python
with sess.graph.as_default():
    tf.import_graph_def(graph_def)
```

```python
input_tensor = sess.graph.get_tensor_by_name('import/input_layer:0')
```

```python
output_tensor = sess.graph.get_tensor_by_name('import/activation_31/Sigmoid:0')
```

```python
output_tensor.shape
```

```python
import numpy as np
```

```python
x = np.zeros((10, 1024, 1024, 3), dtype=np.float32)
```

```python
%%time
y = sess.run(output_tensor, {input_tensor: x})
```
